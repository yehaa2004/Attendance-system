```python
import cv2
import mediapipe as mp
from google.colab.patches import cv2_imshow

# Initialize Mediapipe Hand solutions
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_drawing = mp.solutions.drawing_utils

# Open a webcam feed
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Process the frame and detect hands
    result = hands.process(frame_rgb)

    # Draw hand landmarks on the frame
    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

    # Display the frame
    cv2_imshow(frame)

    # The following line is removed as it's not compatible with Colab
    # if cv2.waitKey(1) & 0xFF == 27:
    #     break

cap.release()
# The following line is removed as it's not compatible with Colab
# cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
import pickle
from google.colab.patches import cv2_imshow

# Load the known face encodings
with open('encodings.pkl', 'rb') as f:
    known_face_encodings, known_face_names = pickle.load(f)

# Open a webcam feed
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    with mp.solutions.face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5) as face_detection:
        results = face_detection.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        if results.detections:
            for detection in results.detections:
                bboxC = detection.location_data.relative_bounding_box
                ih, iw, _ = frame.shape
                (x, y, w, h) = int(bboxC.xmin * iw), int(bboxC.ymin * ih), int(bboxC.width * iw), int(bboxC.height * ih)
                face_roi = gray_frame[y:y+h, x:x+w]

                face_roi_resized = cv2.resize(face_roi, (100, 100))
                face_encoding = face_roi_resized.flatten() / 255.0

                # Compare with known faces
                matches = []
                for known_encoding in known_face_encodings:
                    dist = np.linalg.norm(face_encoding - known_encoding)
                    matches.append(dist)

                name = "Unknown"
                if min(matches) < 0.6:  # Threshold can be adjusted
                    name = known_face_names[np.argmin(matches)]

                # Draw a bounding box and display the name
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
                cv2.putText(frame, name, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

    # Display the frame
    cv2_imshow(frame)

    # The following line is removed as it's not compatible with Colab
    # if cv2.waitKey(1) & 0xFF == 27:
    #     break

cap.release()
```

```python
import matplotlib.pyplot as plt

# Load encodings and names from file
with open('encodings.pkl', 'rb') as f:
    known_face_encodings, known_face_names = pickle.load(f)

# Initialize video capture with a video file
video_path = '/content/known_faces/WhatsApp Video 2025-08-01 at 9.56.33 PM.mp4'
video_capture = cv2.VideoCapture(video_path)

# Initialize Mediapipe face detection
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
face_detection = mp_face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5)

# Initialize attendance list
attendance_log = []

def mark_attendance(name):
    now = datetime.now()
    dt_string = now.strftime('%Y-%m-%d %H:%M:%S')
    attendance_log.append(f"{name}, {dt_string}")
    print(f"Attendance marked for {name} at {dt_string}")

def show_frame(frame):
    # Convert the frame from BGR to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 6))
    plt.imshow(rgb_frame)
    plt.axis('off')
    plt.show()

frame_count = 0
max_frames = 1  # Set a limit to the number of frames to process

while video_capture.isOpened() and frame_count < max_frames:
    ret, frame = video_capture.read()
    if not ret:
        break

    # Convert the frame to RGB
    rgb_frame = frame[:, :, ::-1]

    # Process the frame for face detection
    results = face_detection.process(rgb_frame)

    # Draw face detections
    if results.detections:
        for detection in results.detections:
            mp_drawing.draw_detection(frame, detection)

            # Get the bounding box
            bboxC = detection.location_data.relative_bounding_box
            ih, iw, _ = frame.shape
            (x, y, w, h) = int(bboxC.xmin * iw), int(bboxC.ymin * ih), \
                           int(bboxC.width * iw), int(bboxC.height * ih)

            # Extract the face ROI
            face_roi = cv2.cvtColor(frame[y:y+h, x:x+w], cv2.COLOR_BGR2GRAY)

            # Resize the face ROI to a fixed size
            face_roi_resized = cv2.resize(face_roi, (100, 100))

            # Flatten the face ROI and normalize it
            face_encoding = face_roi_resized.flatten() / 255.0

            # Compare the face with known faces
            distances = np.linalg.norm(np.array(known_face_encodings) - face_encoding, axis=1)
            print(distances)
            min_distance_index = np.argmin(distances)
            print(min_distance_index)
            print(distances[min_distance_index])
            if distances[min_distance_index] < 30:  # Threshold value
                name = known_face_names[min_distance_index]

                # Mark attendance
                if name not in attendance_log:
                    mark_attendance(name)

                # Display the name
                cv2.putText(frame, name, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the resulting frame
    show_frame(frame)

    frame_count += 1

# Release video capture
video_capture.release()
```
